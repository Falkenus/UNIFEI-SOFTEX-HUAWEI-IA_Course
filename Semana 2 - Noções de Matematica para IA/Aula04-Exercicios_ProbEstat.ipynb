(2) Seja $X$ uma variável aleatória contínua com distribuição normal padrão $N(0, 1)$.

```python
# A fdp de X está implementada na biblioteca Scipy

from scipy.stats import norm
```

a) Construa o gráfico da fdp associada à variável $X$ no intervalo $[-6, 6]$.

```python
x = np.linspace(-6, 6, 100)
y = norm.pdf(x)
plt.plot(x, y, 'r-', lw = 2)

plt.fill_between(x, y, color= "r", alpha= 0.2)
plt.grid()
plt.show()
```

Nota: lembrar que no caso da distribuição normal temos

$$
P(a \leq X \leq b) = \phi(b) - \phi(a),
$$

onde

$$
\phi(s) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{s} e^{-x^2/2} dx
$$

e o cálculo pode ser realizado com o comando norm.cdf( )

b) Calcule a probabilidade de que $X$ seja menor ou igual a 1.7, isto é, calcule $P(X \leq 1.7)$.

```python
print(norm.cdf(1.7))
```

c) Calcule $P(1.3 \leq X \leq 1.7)$. \\



```python
prob = norm.cdf(1.7) - norm.cdf(1.3)
print(prob)
```

(3) Considere o seguinte conjunto de dados:

$$
\{1.21, 3.4, 2, 4.66, 1.5, 5.61, 7.22\}.
$$

Calcule a média, a variância e o desvio padrão associados a estes dados.

```python
dados = np.array([1.21, 3.4, 2, 4.66, 1.5, 5.61, 7.22])

media = np.mean(dados)
vari = np.var(dados)
desvio = np.std(dados)

print(media)
print(vari)
print(desvio)
```
