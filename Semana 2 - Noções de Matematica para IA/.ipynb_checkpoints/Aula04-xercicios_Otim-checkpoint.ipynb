# Exercícios resolvidos

Veremos abaixo como resolver o problema de minimizar a função

$$
f(x) = (1 - x_1)^2 + 100(-x_1^2 - x_2)^2 + 1,
$$

utilizando o algoritmo "Newton-CG" implementado na rotina *minimize* da biblioteca Scipy. Esta função tem um único minimizador global dado por $x^* = (1, -1)^T$ e $f(x^*) = 1$.

```python
# Bibliotecas que serão usadas.

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

Para utilizar a rotina, é conveniente implementar $f(x)$, $\nabla f(x)$  e $\nabla^2 f(x)$ separadamente.

```python
# Definindo a função objetivo

def f(x):
  return (1.0 - x[0])**2.0 + 100.0*(-x[0]**2.0 - x[1])**2.0 + 1.0
```

```python
# Definindo o vetor gradiente de f

def gradf(x):
  g = np.zeros_like(x)
  g[0] = -2.0 + 2.0*x[0] + 400.0*x[0]**3.0 + 400.0*x[0]*x[1]
  g[1] = 200.0*x[0]**2.0 + 200.0*x[1]
  return g
```

```python
# Definindo a matriz hessiana de f

def hessf(x):
  n = len(x)
  h = np.zeros((n, n))
  h[0,0] = 2.0 + 1200.0*x[0]**2.0 + 400.0*x[1]
  h[0,1] = 400.0*x[0]
  h[1,0] = 400.0*x[0]
  h[1,1] = 200.0
  return h
```

Abaixo está a sintaxe da rotina *minimize*. O tipo de algoritmo utilizado (no caso "Newton - CG") deve ser especificado dentro da opção *method*. Se o vetor gradiente e a matriz hessiana não estão disponíveis (ou são difíceis de serem calculados), basta tornar jac e hess iguais a "None" que a rotina fará o cálculo usando aproximação por diferenças finitas.

```python
# Chamando a rotina de otimização "Newton - CG"

x0 = np.array([-1.2, 1.0])

sol = minimize(f, x0, method='Newton-CG',
               jac=gradf, hess=hessf,
               options={'xtol': 1e-8, 'disp': True})

print(sol.x)
```

 A figura abaixo ilustra algumas curvas de nível da função $f(x)$. O ponto inicial $x_0$ fornecido para a rotina de otimização e o minimizador global $x^*$ também estão indicados na figura.

```python
N = 1000
xval = np.linspace(-1.3, 1.5, N)
yval = np.linspace(-1.5, 1.4, N)

x0 = np.array([-1.2, 1])
y0 = np.array([1, -1])

X, Y = np.meshgrid(xval, yval)

Z = (1.0 - X)**2.0 + 100.0*(-X**2.0 - Y)**2.0 + 1.0

plt.contour(X, Y, Z, levels = 30)
plt.plot(x0,y0,'r.', ms = 10)
plt.show()
```

# Faça você mesmo!

Considere a função de duas variáveis dada por $f(x) = 2x_1^3 - 3x_1^2 - 6 x_1 x_2 (x_1 - x_2 - 1)$. Partindo de $x_0 = (5, 5)^T$ aplique a rotina de otimização *minimize* para encontrar um minimizador local de $f(x)$, fornecendo o vetor gradiente e a matriz hessiana verdadeiros. Coloque "Newton-CG" como opção para a rotina. Após a resolução do problema, construa algumas curvas de nível de $f(x)$ e indique no mesmo gráfico o ponto inicial $x_0$ e a solução encontrada $x^*$. 

```python
def f(x):
  return 2.0*x[0]**3 - 3.0*x[0]**2 - 6.0*x[0]*x[1]*(x[0]-x[1]-1)

def gradf(x):
  g = np.zeros_like(x)
  g[0] = 6.0*x[0]**2 - 6.0*x[0] - 12.0*x[0]*x[1] + 6.0*x[1]**2 + 6.0*x[1]
  g[1] = -6.0*x[0]**2 + 12.0*x[0]*x[1] + 6*x[0]
  return g

def hessf(x):
  n = len(x)
  h = np.zeros((n, n))
  h[0,0] = 12.0*x[0] - 6.0 - 12.0*x[1]
  h[0,1] = -12.0*x[0]  + 12.0*x[1] + 6
  h[1,0] = -12.0*x[0]  + 12.0*x[1] + 6
  h[1,1] = 12.0*x[0]
  return h
                                                                                 
x0 = (np.array([5.0, 5.0]))

sol = minimize(f, x0, method='Newton-CG',
               jac=gradf, hess=hessf,
               options={'xtol': 1e-8, 'disp': True})

print(sol.x)
                                                                                 
N = 1000
xval = np.linspace(-1.3, 1.5, N)
yval = np.linspace(-1.5, 1.4, N)

x0 = np.array([-1.2, 1])
y0 = np.array([1, -1])

X, Y = np.meshgrid(xval, yval)

Z = 2*X**3 - 3*X**2 - 6*X*Y*(X - Y - 1)

plt.contour(X, Y, Z, levels = 30)
plt.plot(x0,y0,'r.', ms = 10)
plt.show()
```
